# WEEK 2: Data Modeling

```python
x=df.drop(['Outcome'],axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

# Compare various models with the results from KNN algorithm

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,roc_curve,auc
```

```python
#creating the objects
lr = LogisticRegression(solver='liblinear',random_state=42)
nb=GaussianNB()
sgd=SGDClassifier(random_state=42)
knn=KNeighborsClassifier()
dtc=DecisionTreeClassifier(random_state=42)
rf=RandomForestClassifier(random_state=42)
svc=SVC(kernel='linear',random_state=42)

dict = {0: 'Logistic Regression', 1: 'GaussianNB',2:'SGDClassifier',3:'KNeighborsClassifier',4:'DecisionTreeClassifier',5:'RandomForestClassifier',6:'SVC'}
models=[lr,nb,sgd,knn,dtc,rf,svc]


for i,model in enumerate(models):
    print("{} Test Accuracy: {}".format(dict[i],cross_val_score(model, x_train,y_train, cv=10, scoring ='accuracy').mean()))
```

```python
#Preparing ROC Curve (Receiver Operating Characteristics Curve)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#logistic regression 
lr.fit(x_train,y_train)
# predict probabilities
probs = lr.predict(x_train)
# keep probabilities for the positive outcome only
#probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(y_train, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_train, probs)
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
```
